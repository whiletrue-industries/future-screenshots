```python
# PYTHON
# Take 10000 x, y from a uniform distribution ^ 2
import numpy as np
num = 10000
x = 1 - np.random.uniform(0, 1, 10000) ** 0.5
y = 1 - np.random.uniform(0, 1, 10000) ** 0.5

# draw the result as a scatter plot:
import matplotlib.pyplot as plt

plt.scatter(x, y, alpha=0.5)
plt.title('Scatter plot of random points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
